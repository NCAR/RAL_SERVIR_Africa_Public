# Functions

```python
# Function to calculate air density

def calculate_air_density(sp, q, t):
    p_dry = sp * (1 - q * 28.96 / 18)
    R = 287.0500676
    air_density = p_dry / (R * t)
    air_density = air_density.rename({'sp': 'air_d'})
    return air_density



def calculate_specific_humidity(T, T_d, sp):
    # Calculate water vapor pressure (e) in mb
    e = 6.112 * math.exp((17.67 * T_d) / (T_d + 243.5))

    # Calculate specific humidity (q) in kg/kg
    specific_humidity = (0.622 * e) / (sp - 0.378 * e)
    return specific_humidity
```

# Reading data

```python
# reading data
file_name_data=glob(cams_data_dir+'*')
file_name_data.sort()
file_name_data
```

```python
%%time
if files_separation=='variables':
    air_q=xr.open_mfdataset(file_name_data)
elif files_separation=='temporal':
    air_q=xr.open_mfdataset(file_name_data, combine='nested',  concat_dim='time')

coordinate_names=list(air_q.coords)
lev_name = [ coord for coord in coordinate_names if 'lev' in coord]

air_q=air_q.sortby('time')
display(air_q)
```

```python
air_q[species]
```

```python
print('\n ****** Detailed look at the metadata ******')
display(air_q.attrs)

#saving the input attributes for final saved file
input_attributes = {}
input_attributes['global']  =air_q.attrs
```

# Unit conversions

```python
%%time
if species=='pm2p5':
    concentration_daily=1e9 * air_q[species]
    concentration_daily.attrs['units']= 'ug/m^3'
else:        
    concentration_daily = 1e9 * air_q[species] * 28.96 / mole_weight
    concentration_daily.attrs['units'] = 'ppbv'
    

#convert_longitude_from_0_360 to -180 to 180
concentration_daily['longitude'] = xr.where(concentration_daily['longitude'] > 180, concentration_daily['longitude'] - 360, concentration_daily['longitude'])
concentration_daily = concentration_daily.sortby('longitude')
concentration_daily = concentration_daily.sortby('latitude')
concentration_daily

# add global attributes from source data
concentration_daily.attrs['Attributes']=str(input_attributes['global'])
concentration_daily.name = species
del air_q
concentration_daily
```

# Estimate seasonal average over Africa

set your plot input

```python
#set your input
season_str=['winter','winter']
year=[2005,2015]

# cropping Africa based on latitude and longitude bounds
min_lat,max_lat=-40, 40
min_lon,max_lon=-20, 50
colormap='plasma'
```

```python
%%time
# seasonal mean
seasonal_data = concentration_monthly.resample(time='QS-DEC').mean()


# cropping Africa based on latitude and longitude bounds
seasonal_data_cropped_africa = seasonal_data.sel(latitude=slice(min_lat,max_lat), longitude=slice(min_lon,max_lon))
seasonal_data_cropped_africa
del seasonal_data
seasonal_data_cropped_africa
```

```python
%%time
#########

season_dict = {
    'winter': 12,
    'spring': 3,
    'summer': 6,
    'autumn': 9
}
sel_times = ['{}-{}'.format(y, season_dict[s]) for y, s in zip(year, season_str)]
for idx, (time_step) in enumerate(sel_times):
    # Create a map plot
    fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(10, 10))
    if   lev_name: 
        time_step_data = seasonal_data_cropped_africa.sel(time=time_step,level=lv_sel)
    else:
        time_step_data = seasonal_data_cropped_africa.sel(time=time_step) 
    time_step_data.plot(ax=ax, cmap=colormap, vmin=vmin_set,vmax=vmax_set)
    
    # Set the extent to focus on Africa
    ax.set_extent([min_lon,max_lon,min_lat,max_lat], crs=ccrs.PlateCarree())
    ax.coastlines()
    ax.add_feature(cartopy.feature.BORDERS, linestyle=':', linewidth=0.7)
    
    # adding the axis ticks
    ax.set_xticks(np.arange(min_lon,max_lon, 20), crs=ccrs.PlateCarree())
    ax.set_yticks(np.arange(min_lat,max_lat, 20), crs=ccrs.PlateCarree())
 
    plt.ylabel('Latitude')
    plt.title('{} seasonal average {} {}'.format(species_str,season_str[idx],year[idx]))
    fig.set_size_inches(8.5, 10.5)  

    #save fig
    save_name='{}_Africa_season_ave_{}_{}'.format(species,season_str[idx],year[idx])
    plt.savefig(output_path_plots+save_name,dpi=300)
del seasonal_data_cropped_africa
```

# Plotting time series of annual concentrations for the locations of interest

set your plot input

```python
cities = {
  "Nairobi": [-1.2921, 36.8219],
  #"Kigali": [-1.9441, 30.0619],
  # "Kampala": [0.3476, 32.5825],
  # "Dodoma": [-6.1630, 35.7516],
  # "Lilongwe": [-13.9626, 33.7741],
  # "Lusaka": [-15.3875, 28.3228]
}

annual_period=['2005','2020']
```

```python
%%time
# annual mean
annual_data = concentration_monthly.resample(time='1Y').mean()
annual_data.time

if   lev_name:      
    time_s_cons=annual_data.sel(time=slice(*annual_period),level=lv_sel)
else:
    time_s_cons=annual_data.sel(time=slice(*annual_period))
```

```python
%%time

for idx, (city, coordinates) in enumerate(cities.items()):
    fig,ax=plt.subplots()
    lat,lon=coordinates
    print(f"City: {city}, Latitude: {lat}, Longitude: {lon}")
    nearest_location = time_s_cons.sel(latitude=lat, longitude=lon, method='nearest')
    nearest_location.plot(label=city, c='k',marker='o', linestyle='-')
    title_str=species+'({})'.format(time_s_cons.attrs['units'])
    # Calculate the trend line 
    time_val=nearest_location.time.values
    time_arr=np.linspace(1,len(nearest_location.time),len(nearest_location.time))
    coefficients =np.polyfit(time_arr,nearest_location.values,deg=1)
    trend_line = np.polyval(coefficients, time_arr)

    # Create the trend line
    equation = f'Trend Line: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}'
    plt.plot(time_val,trend_line,label=equation, linestyle='--', color='red') 
    #legend, title, and plot size set up
    ax.set_title('annual {} for {}'.format(species_str,city))
    ax.legend()    
    fig.set_size_inches(10,4)
    save_name='Annual_{}_{}_{}_{}'.format(species, annual_period[0], annual_period[1],city)
    plt.savefig(output_path_plots+save_name+'.jpg',dpi=300)
```
